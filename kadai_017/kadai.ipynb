```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")

humans = []

humans.append(Human("侍太郎", 36))
humans.append(Human("侍一郎", 20))
humans.append(Human("侍二郎", 18))

for human in humans:
  human.check_adult()
```
